```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

print("NAME: SUBASRI R 24BAD116")
df = pd.read_csv("data.csv", encoding="ISO-8859-1")

print(df.head())
print(df.tail())
print(df.info())
print(df.describe())

print(df.isnull().sum())

df['Sales'] = df['Quantity'] * df['UnitPrice']

product_sales = df.groupby('Description')['Sales'].sum().head(10)

product_sales.plot(kind='bar')
plt.title("Top 10 Products by Sales")
plt.ylabel("Total Sales")
plt.show()

product_sales.plot(kind='line', marker='o')
plt.title("Sales Trend of Top Products")
plt.ylabel("Total Sales")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("NAME: SUBASRI R 24BAD116")
df = pd.read_csv("diabetes.csv")
print(df.head())
print(df.info())
print(df.isnull().sum())
cols = ['Glucose', 'BloodPressure', 'BMI', 'Insulin']
df[cols] = df[cols].replace(0, np.nan)
print(df.isnull().sum())
plt.hist(df['Glucose'].dropna(), bins=20)
plt.title("Glucose Level Distribution")
plt.xlabel("Glucose")
plt.ylabel("Frequency")
plt.show()
sns.boxplot(x=df['Age'])
plt.title("Age Distribution")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("NAME: SUBASRI R 24BAD116")
df = pd.read_csv("Housing.csv")
print("First 5 rows:")
print(df.head())
print("\nDataset Information:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
plt.figure()
plt.scatter(df['area'], df['price'])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
numeric_df = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap (Numerical Features)")
plt.show()
print("\nObservations:")
print("- Area shows strong positive correlation with house price.")
print("- Number of bathrooms and bedrooms influence price.")
print("- Parking availability also increases house value.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print("NAME: SUBASRI R 24BAD116")
df = pd.read_csv("marketing_campaign.csv", sep='\t')
print(df.head())
print(df.info())
print(df.isnull().sum())
df['Age'] = 2024 - df['Year_Birth']
sns.histplot(df['Age'], bins=20)
plt.title("Age Distribution")
plt.show()
sns.boxplot(x=df['Income'])
plt.title("Income Distribution")
plt.show()
df[['MntWines', 'MntFruits', 'MntMeatProducts']].sum().plot(kind='bar')
plt.title("Customer Spending Pattern")
plt.ylabel("Amount Spent")
plt.show()
```
